```python
import numpy as np
import pandas as pd
bk = pd.read_csv('books.csv')
us = pd.read_csv('users.csv')
rt = pd.read_csv('ratings.csv')
bk.head()
```

## popularity recomender



```python
bk_rt= rt.merge(bk,on='ISBN')
bk_rt
```

```python
count_rt = bk_rt.groupby("Book-Title").count()["Book-Rating"].reset_index()
count_rt.rename(columns={"Book-Rating":"count_rt"},inplace =True)
count_rt
```

```python
avg_rt = bk_rt.groupby('Book-Title') ['Book-Rating'].mean() .reset_index()
avg_rt.rename(columns={"Book-Rating":"avg_rt"},inplace =True)
avg_rt
```

```python
pop_rs = avg_rt.merge(count_rt, on='Book-Title')
pop_rs
```

```python
pop_rs = pop_rs[pop_rs['count_rt']>=200].sort_values('avg_rt',ascending = False).head(50)
pop_rs = pop_rs
```

```python
pop_rs=pop_rs.merge(bk,on ='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','count_rt','avg_rt']]
pop_rs
```

## colabrative recomender


```python
x=bk_rt.groupby('User-ID').count()['Book-Rating']>200 
kn_us=x[x].index
```

```python
filter_rt = bk_rt[bk_rt['User-ID'].isin(kn_us)]
filter_rt
    
```

```python
y = filter_rt.groupby("Book-Title").count()["Book-Rating"]>=50
fam = y[y].index
print(fam)
```

```python
final= filter_rt[filter_rt["Book-Title"].isin(fam)]
final.drop_duplicates("Book-Title")
```

```python
pt=final.pivot_table(index = "Book-Title",columns = "User-ID", values = "Book-Rating" )
pt.fillna(0,inplace = True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
sim =  cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index== book_name)[0][0]
    sim_it = sorted(list(enumerate(sim[index])),key = lambda x:x[1],reverse = True)[1:6]
    
    #printing the suggested book
    for i in sim_it:
        print (pt.index[i[0]])
```

```python
recommend("2nd Chance")
```
